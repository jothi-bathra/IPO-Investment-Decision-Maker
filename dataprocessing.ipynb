```python
df_final = pd.DataFrame(columns = ['Name' , 'City', 'Country' , 'Industry', 'Sector', 'Size'])
```

```python
for company in tickers:
    symbol = company
    GetInformation = yahooFinance.Ticker(symbol)
    name=symbol
    city=GetInformation.info['city']
    country=GetInformation.info['country']
    industry=GetInformation.info['industry']
    sector=GetInformation.info['sector']
    size=GetInformation.info['fullTimeEmployees']
    data={'Name':name,'City':city,'Country':country,'Industry':industry,'Sector':sector,'Size':size}
    new_row_df = pd.DataFrame([data])
    df_final = pd.concat([df_final, new_row_df], ignore_index=True)
    
```
